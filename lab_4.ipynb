```python
import numpy as np
n = 8
points = np.array([np.random.uniform(low = 0, high = 10,size = 3) for _ in range(n)])
points = np.concatenate((points, np.ones(shape = [n, 1])), axis = 1)
```

```python
points
```

```python
T1 = np.array([[0, 0, 0]]).T
T2 = np.array([[np.random.uniform(low = 1, high = 10),0,0]]).T
```

```python
E = np.diag(np.ones((3, )))
P1  = np.concatenate((E, T1), axis = 1)
P2 = np.concatenate((E, T2), axis = 1)
```

```python
first_img_points = np.dot(P1, points.T).T
second_img_points = np.dot(P2, points.T).T
```

```python
first_img_points
```

```python
second_img_points
```

```python
res = []
for p1, p2 in zip(first_img_points, second_img_points):
    res.append([
        p1[0] * p2[0],
        p1[1] * p2[0],
        p2[0],
        p1[0] * p2[1],
        p1[1] * p2[1],
        p2[1],
        p1[0],
        p1[1],
        1])
```

```python
res = np.array(res).T
```

```python
U, S, Vt =  np.linalg.svd(res)
```
