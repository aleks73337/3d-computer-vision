```python
def func(f_x, f_y, c_x, c_y, a, a_x, a_y, a_z, x_0, y_0, z_0):
    K = np.array([[f_x, 0, c_x], [0, f_y, c_y], [0, 0, 1]])
    Mz = np.array([[np.cos(a), -np.sin(a), 0], [np.sin(a), np.cos(a), 0], [0, 0, 1]])
    T = np.array([a_x, a_y, a_z]).reshape(3, 1)
#     print(np.concatenate())
    P = np.matmul(K, np.concatenate((Mz, T), axis = 1))
    result = np.matmul(P, np.array([x_0, y_0, z_0, 1]))
    return result, K

result, K = func(f_x = 400, f_y = 400, c_x = 960, c_y = 540, a = 45, a_x = 0, a_y = 0, a_z = 10, x_0 = 10, y_0 = -10, z_0 = 100)
```

```python
def func2(x1, y1, xp1, yp1, x2, y2, xp2, yp2, x3, y3, xp3, yp3, x4, y4, xp4, yp4):
    A=[
    [-x1,  -y1,  -1,   0,    0,    0,   x1*xp1,   y1*xp1,   xp1],
    [0,    0,    0, -x1,   -y1,  -1,   x1*yp1,   y1*yp1,   yp1],
    [-x2,  -y2,  -1,   0,    0,    0,   x2*xp2,   y2*xp2,   xp2],
    [0,    0,    0, -x2,   -y2,  -1,   x2*yp2,   y2*yp2,   yp2],
    [-x3,  -y3,  -1,   0,    0,    0,   x3*xp3,   y3*xp3,   xp3],
    [0,    0,    0, -x3,   -y3,  -1,   x3*yp3,   y3*yp3,   yp3],
    [-x4,  -y4,   -1,  0,    0,    0,   x4*xp4,   y4*xp4,   xp4],
    [0,    0,    0,  -x4,  -y4,  -1,   x4*yp4,   y4*yp4,   yp4]]
    np.array(A)
    U, S, V = np.linalg.svd(A)
    res = V[:, -1].reshape(3,3)
    return res
```
